Missing values
There are different causes for missing values: A value might not be filled in, or simply does not exist (e.g. no surface for room 2 for a 1 room house)
Three approaches to deal with missing values:
1. Drop columns with missing values, simple but we might be losing a lot of useful information
2. Imputation: filling in missing values, e.g. filling in an entry of a column by taking the average of the column
3. Imputation and tracking of imputation: Apply imputation, but also keep track of what values where imputed in a new column

Categorical variables
Only take on a limited number of values. Categorical variables should be preprocessed before inserting into most machine learning models
1. Dropping columns with categorical variables -> only works when the column doesn't contain useful information
2. Oridal encoding -> assigning an integer per unique category, work wel with ordinal variables, variables with an order: bad, normal, good, best
3. One-hot encoding -> Creates new columns indicating the presence of the category (binary). Used when there is no intrinsic ranking among the categories, e.g. colors: blue, red, green. one-hot encoding does not work wel if the category can take on many values (e.g. > 15)

Data leakage
Training data contains information about the target, but similar data will not be available when the model is used for prediction

Target leakage occurs when the predictors include data that will not be available at the time of making predictions, i.e. the data changes between those moments. To prevent target leakage, any variable updated (or created) after the target value is realized should be excluded

Train-test contamination, just mind that your validation set is not included in any fitting step

Example with credict card applications: we are misleaded by the crossvalidation results of 98% accuracy because the training data set contains columns that contain information derived after the application was accepted, normally this data would not be available
